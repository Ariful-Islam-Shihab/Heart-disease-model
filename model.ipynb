```python
import pandas as pd
import joblib
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
```

```python
dataset=pd.read_csv('/content/heart.csv')
dataset.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.countplot(x='target', data=dataset)
plt.title('Outcome Count')
plt.show()
```

```python
dataset.describe()
```

```python
X_train, X_test, y_train, y_test = train_test_split(dataset.drop(['target','restecg','exang','slope','ca','thal'], axis=1), dataset['target'], test_size=0.2, random_state=42)
```

```python
X_train.shape, X_test.shape
```

```python
rf_classifier=RandomForestClassifier(n_estimators=120,random_state=62,max_depth=4)
rf_classifier.fit(X_train,y_train)
y_pred=rf_classifier.predict(X_test)
reprt=classification_report(y_test,y_pred)
print(reprt)
```

```python
joblib.dump(rf_classifier, 'heart_disease_model.pkl')
```
